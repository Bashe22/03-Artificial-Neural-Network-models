### Train Test Split

```python
# let's first seperate the label column from features
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
# importing sicket-learn's trainig_test_split
from sklearn.model_selection import train_test_split

# splitting the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```


### Scaling Data

```python
# importing sicket-learn's MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# creating the scaller and fitting it
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# importing tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# double checking training and test data shape
print(X_train.shape)
print(X_test.shape)
```

```python
# creating sequential model dropout layers 
model = Sequential()

# adding input hidden layers
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))

# adding output layer and since this is binary classification activation will be sigmoid
model.add(Dense(units=1,activation='sigmoid'))

# compiling the model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
# adding early stopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

### Training the Model

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

Fantastic! both training loss and validation loss goes down and flatten out at the same rate.

### Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix

predictions = model.predict_classes(X_test)

# printing our classification report against predictions
print(classification_report(y_test,predictions))
```

The report shows very good performance and we can confirm it using confusion matrix

```python
print(confusion_matrix(y_test,predictions))
```
